# Image Segmentation with Sobel-Based Gradients

This notebook demonstrates the use of the **Safas** filter script **sobel_focus**. 

## Import packages and data

In addition to **safas** import packages required to load and display results. 

```python
import sys
sys.path.append(r'C:\Users\Ryan\Desktop\camera_setup\safas-master')
# Pacakge for viewing the images
from matplotlib import pyplot as plt
%matplotlib inline

import safas as sfs
from safas import data
from safas.filters.sobel_focus import imfilter as sobel_filter
```

## Process sample flocs image with threshold and sobel-based filter

Multiple steps are often requires to extract the object of interest from the image. A good introduction to image segmentation may be read in the [sci-kit image](https://scikit-image.org/docs/dev/user_guide/tutorial_segmentation.html) or [OpenCV](https://docs.opencv.org/master/d2/d96/tutorial_py_table_of_contents_imgproc.html) documentation.

```python
img = data.mudflocs()

# apply threshold to segment the image 
params = {'img_thresh': 90,
          'focus_filter': False,
          'clearborder_filter': False,
          'contour_color': (0,255,0)}
        
thresh_labels, thresh_contours = sobel_filter.imfilter(src=img.copy(), **params)

# apply sobel-based gradient to segment the image 
params = {'img_thresh': 70,
          'edge_thresh': 50,
          'focus_filter': True,
          'clearborder_filter': True,
          'contour_color': (0,0,255),}
        
sobel_labels, sobel_contours = sobel_filter.imfilter(src=img.copy(), **params)

f, ax  = plt.subplots(1,3, dpi=250, figsize=(7.5, 3.5))
ax = ax.ravel()
for a in ax: 
    a.set_xticks([])
    a.set_yticks([])
    
ax[0].imshow(img)
ax[0].set_title('Raw image')
ax[1].imshow(thresh_contours)
ax[1].set_title('Threshold filter')
ax[2].imshow(sobel_contours)
ax[2].set_title('Sobel-based filter')
plt.tight_layout()
```
